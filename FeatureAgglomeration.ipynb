```python
from sklearn.cluster import FeatureAgglomeration
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = np.load('byte_prob_distr_no_zero.npy')
agglo = FeatureAgglomeration(n_clusters=3).fit_transform(data)

```

```python
fig = plt.figure()
#ax = Axes3D(fig)
plt.scatter(agglo[:,0], agglo[:,1])
plt.show()
```
